## PART 0. DOWNLOADING AND PREPARING THE DATA

We import the necessary libraries

```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

Download the data and read it into a pandas dataframe

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

Check for missing values

```python
concrete_data.describe()
```

```python
concrete_data.isnull().sum()
```

Our dataset looks clean. Now we can start to build our model.

# PART 1

We first split data into predictors and target. Then we divide into training and test set.

```python
# split into predictors and target
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column

 # number of predictors, we're gonna need it later
n_cols = predictors.shape[1]

# train test split
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)

```

We define the model.

```python
def regression_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

Build and fit the model.

```python
# build
model = regression_model()
# fit
m = model.fit(X_train, y_train,epochs=50, verbose=2)
```

We calculate the mean square error.

```python
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
```

```python
print('The mean of the mean squared errores is %s' %"{:.2f}".format(mse))
```

Repeat the process 50 times.

```python
mseList = []

for i in range(1,51):
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)
    model = regression_model()
    m = model.fit(X_train, y_train,epochs=50, verbose=2)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mseList.append(mse)


aux = np.array(mseList)


mseMeanPart1 = aux.mean()
stdMeanPart1 = aux.std()
```

```python
print('The mean of the mean squared errores is %s' %"{:.2f}".format(mseMeanPart1))
print('The standard deviation of the mean squared errores is %s' %"{:.2f}".format(stdMeanPart1))
```

# PART 2

Now, we normalize the data.

```python
mseListNormalizedData = []

for i in range(1,51):
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)
    # normalize
    X_train_norm = (X_train - X_train.mean() / X_train.std())
    X_test_norm = (X_test - X_test.mean() / X_test.std())
    # build, fit and error
    model = regression_model()
    m = model.fit(X_train_norm, y_train,epochs=50, verbose=2)
    y_pred = model.predict(X_test_norm)
    mse = mean_squared_error(y_test, y_pred)
    mseListNormalizedData.append(mse)


aux2 = np.array(mseListNormalizedData)

mseMeanPart2 = aux2.mean()
stdMeanPart2 = aux2.std()
```

```python
print('The mean of the mean squared errores is %s' %"{:.2f}".format(mseMeanPart2))
print('The standard deviation of the mean squared errores is %s' %"{:.2f}".format(stdMeanPart2))
```

We see that the error and the standard deviation decreased considerably compared to part one

# PART 3

We increment the number of epochs.

```python
listPart3 = []
epochs = 100

for i in range(1,51):
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)
    X_train_norm = (X_train - X_train.mean() / X_train.std())
    X_test_norm = (X_test - X_test.mean() / X_test.std())
    model = regression_model()
    m = model.fit(X_train_norm, y_train,epochs=epochs, verbose=2)
    y_pred = model.predict(X_test_norm)
    mse = mean_squared_error(y_test, y_pred)
    listPart3.append(mse)


aux3 = np.array(listPart3)

mseMeanPart3 = aux3.mean()
stdMeanPart3 = aux3.std()


```

```python
print('The mean of the mean squared errores is %s' %"{:.2f}".format(mseMeanPart3))
print('The standard deviation of the mean squared errores is %s' %"{:.2f}".format(stdMeanPart3))
```

We see that the error and the standard deviation decreased considerably compared to part two

# PART 4

We add hidden layers.

```python
def regression_model2():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))       
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
listPart4 = []
epochs = 50

for i in range(1,51):
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)
    # normalize
    X_train_norm = (X_train - X_train.mean() / X_train.std())
    X_test_norm = (X_test - X_test.mean() / X_test.std())
    # construimos el modelo
    model = regression_model2()
    # ajustamos el modelo
    m = model.fit(X_train_norm, y_train,epochs=epochs, verbose=2)
    y_pred = model.predict(X_test_norm)
    mse = mean_squared_error(y_test, y_pred)
    listPart4.append(mse)


aux4 = np.array(listPart4)

mseMeanPart4 = aux4.mean()
stdMeanPart4 = aux4.std()
```

```python
print('The mean of the mean squared errores is %s' %"{:.2f}".format(mseMeanPart4))
print('The standard deviation of the mean squared errores is %s' %"{:.2f}".format(stdMeanPart4))
```

We see that the error and the standard deviation decreased considerably compared to part two
